```python
races = {'season':[],
        'round':[],
        'circuit_id':[],
        'lat':[],
        'long':[],
        'country':[],
        'date':[],
         'FirstPractice':[],
         'SecondPractice':[],
         'ThirdPractice':[],
         'Qualifying':[],
        'url':[]}
```

```python
for year in list(range(2010,2023)):
    url='http://ergast.com/api/f1/{}.json'
    response=requests.get(url.format(year))
    json=response.json()
    
    for item in json['MRData']['RaceTable']['Races']:
        
        races['season'].append(int(item['season']))
        races['round'].append(int(item['round']))
        races['circuit_id'].append(item['Circuit']['circuitId'])
        races['lat'].append(item['Circuit']['Location']['lat'])
        races['long'].append(item['Circuit']['Location']['long'])
        races['country'].append(item['Circuit']['Location']['country'])
        races['date'].append(item['date'])
        races['url'].append(item['url'])
        
        try:
            races['FirstPractice'].append(item['FirstPractice']['date'])
            
        except:
            races['FirstPractice'].append(None)
        
        try:
            races['SecondPractice'].append(item['SecondPractice']['date'])
            
        except:
            races['SecondPractice'].append(None)
        
        try:
            races['ThirdPractice'].append(item['ThirdPractice']['date'])
            
        except:
            races['ThirdPractice'].append(None)
        
        try:
            races['Qualifying'].append(item['Qualifying']['date'])
            
        except:
            races['Qualifying'].append(None)
```

```python
df_results_2022=pd.DataFrame()
for race in list(range(1,23)):
    
    url='http://ergast.com/api/f1/2022/{}/results.json'
    response=requests.get(url.format(race))
    content=json.loads(response.content)
    
    df_results_2022=df_results_2022.append(pd.DataFrame(content['MRData']['RaceTable']['Races'][0]['Results']))
```

```python
df_results_2022['Driver']=df_results_2022['Driver'].apply(lambda x: x['driverId'])
```

```python
df_results_2022['Constructor']=df_results_2022['Constructor'].apply(lambda x: x['constructorId'])
```

```python
df_results_2022['points']=df_results_2022['points'].astype(int)
df_results_2022['position']=df_results_2022['position'].astype(int)
```

```python
df_results_2022.groupby(['Driver']).agg({'points':'sum'}).sort_values(by='points', ascending=False) 
```

```python
df_results_2022.groupby(['Constructor']).agg({'points':'sum'}).sort_values(by='points', ascending=False)
```
